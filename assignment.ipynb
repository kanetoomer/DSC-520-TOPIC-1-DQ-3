```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
np.random.seed(0)

# My sample dataset
data = {
    'VoterID': np.arange(1, 101),
    'Name': [f'Voter{i}' for i in range(1, 101)],
    'Age': np.random.randint(15, 90, size=100)  # Ages between 15 and 89
}

# DataFrame
df = pd.DataFrame(data)

# Filter out voters who are under the age of 18
df_filtered = df[df['Age'] >= 18]

# Create histograms
fig, ax = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# Histogram for the original dataset
ax[0].hist(df['Age'], bins=15, color='skyblue', edgecolor='black')
ax[0].set_title('Original Age Distribution')
ax[0].set_xlabel('Age')
ax[0].set_ylabel('Frequency')

# Histogram for the filtered dataset
ax[1].hist(df_filtered['Age'], bins=15, color='lightgreen', edgecolor='black')
ax[1].set_title('Filtered Age Distribution (18 and above)')
ax[1].set_xlabel('Age')

# Display the histograms
plt.tight_layout()
plt.show()
```
